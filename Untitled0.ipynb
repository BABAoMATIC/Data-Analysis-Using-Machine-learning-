```python
import pandas as pd

# Load the file
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MST575 /cell2cell-duke univeristy.csv')

# Check basic info
df.info()

# Preview first few rows
df.head()
```

```python
# Check missing values
df.isnull().sum()

# Drop columns with too many missing or irrelevant
df = df.drop(['Unnamed: 0', 'X', 'customer', 'churndep'], axis=1)

# Fill missing numeric columns with median
df = df.fillna(df.median(numeric_only=True))
```

```python
# Target
y = df['churn']

# Features
X = df.drop('churn', axis=1)
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
!pip install imbalanced-learn

from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_sm, y_train_sm = smote.fit_resample(X_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

lr = LogisticRegression(max_iter=1000)
lr.fit(X_train_sm, y_train_sm)
y_pred_lr = lr.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print("Precision:", precision_score(y_test, y_pred_lr))
print("Recall:", recall_score(y_test, y_pred_lr))
print("F1 Score:", f1_score(y_test, y_pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train_sm, y_train_sm)
y_pred_dt = dt.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred_dt))
print("Precision:", precision_score(y_test, y_pred_dt))
print("Recall:", recall_score(y_test, y_pred_dt))
print("F1 Score:", f1_score(y_test, y_pred_dt))
```

```python
!pip install xgboost

import xgboost as xgb

xgb_clf = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_clf.fit(X_train_sm, y_train_sm)
y_pred_xgb = xgb_clf.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred_xgb))
print("Precision:", precision_score(y_test, y_pred_xgb))
print("Recall:", recall_score(y_test, y_pred_xgb))
print("F1 Score:", f1_score(y_test, y_pred_xgb))
```

```python
# Import SVC first
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Sample 30k rows from training data
X_train_sample = X_train_sm.sample(n=30000, random_state=42)
y_train_sample = y_train_sm.loc[X_train_sample.index]

# Fit SVM on this smaller sample
svm_clf = SVC(kernel='rbf', C=1.0, gamma='scale', probability=True)
svm_clf.fit(X_train_sample, y_train_sample)

# Predict on test data
y_pred_svm = svm_clf.predict(X_test)

# Evaluate
print("SVM Model Performance on 30k Sample:")
print("Accuracy:", accuracy_score(y_test, y_pred_svm))
print("Precision:", precision_score(y_test, y_pred_svm))
print("Recall:", recall_score(y_test, y_pred_svm))
print("F1 Score:", f1_score(y_test, y_pred_svm))
```

**Tunning**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

param_grid_lr = {
    'C': [0.01, 0.1, 1, 10, 100]
}

lr = LogisticRegression(max_iter=1000)
grid_lr = GridSearchCV(lr, param_grid_lr, cv=5, scoring='accuracy', n_jobs=-1)
grid_lr.fit(X_train_sm, y_train_sm)

print("Best LR params:", grid_lr.best_params_)
print("Best LR accuracy:", grid_lr.best_score_)

y_pred_lr = grid_lr.predict(X_test)
print("Test Accuracy:", accuracy_score(y_test, y_pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

param_grid_dt = {
    'max_depth': [5, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'criterion': ['gini', 'entropy']
}

dt = DecisionTreeClassifier(random_state=42)
grid_dt = GridSearchCV(dt, param_grid_dt, cv=5, scoring='accuracy', n_jobs=-1)
grid_dt.fit(X_train_sm, y_train_sm)

print("Best DT params:", grid_dt.best_params_)
print("Best DT accuracy:", grid_dt.best_score_)

y_pred_dt = grid_dt.predict(X_test)
print("Test Accuracy:", accuracy_score(y_test, y_pred_dt))
```

```python
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV

param_dist_xgb = {
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5, 10, 15],
    'n_estimators': [100, 200, 300],
    'subsample': [0.5, 0.7, 1.0]
}

xgb_clf = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
random_search_xgb = RandomizedSearchCV(xgb_clf, param_distributions=param_dist_xgb,
                                       n_iter=10, cv=5, scoring='accuracy', n_jobs=-1)
random_search_xgb.fit(X_train_sm, y_train_sm)

print("Best XGBoost params:", random_search_xgb.best_params_)
print("Best XGBoost accuracy:", random_search_xgb.best_score_)

y_pred_xgb = random_search_xgb.predict(X_test)
print("Test Accuracy:", accuracy_score(y_test, y_pred_xgb))
```

```python
# Copy original DataFrame so we don't lose anything
df_tableau = df.copy()

# 1️⃣ Create Churn Label column
df_tableau['Churn_Label'] = df_tableau['churn'].map({0: 'Not Churned', 1: 'Churned'})

# 2️⃣ OwnRent to Yes/No
df_tableau['OwnRent_Label'] = df_tableau['ownrent'].map({0: 'Own', 1: 'Rent'})

# 3️⃣ MarryYes to Yes/No
df_tableau['MarryYes_Label'] = df_tableau['marryyes'].map({0: 'No', 1: 'Yes'})

# 4️⃣ Optional rounding for prettier visuals
df_tableau['Revenue'] = df_tableau['revenue'].round(2)
df_tableau['Mou'] = df_tableau['mou'].round(2)
df_tableau['Age1'] = df_tableau['age1'].round(0)

# 5️⃣ Drop unnecessary columns
df_tableau = df_tableau.drop(columns=[
    'Unnamed: 0', 'X', 'customer', 'churndep', 'ownrent', 'marryyes', 'churn'  # we keep Churn_Label instead
], errors='ignore')

# Final check
df_tableau.head()

# Export to CSV for Tableau
df_tableau.to_csv('Cell2Cell_for_Tableau.csv', index=False)
```
